```python
from plotnine import *
from plotnine.data import diamonds
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = diamonds
df
```

```python
p = (ggplot(df) + aes(x = "price", y = 'carat') + geom_line())
p.draw()
plt.show()
```

```python
p = (ggplot(df) + aes(x = "price", y = 'carat') + geom_point())
p.draw()
plt.show()
```

```python
p = (ggplot(df) + aes(x = "price", y = 'carat', color = 'cut') + geom_point())
p.draw()
plt.show()
```

```python
p = (ggplot(df) + aes(x = "cut", y = 'price', fill = 'cut') + geom_bar(stat = 'identity'))
p.draw()
plt.show()
```

```python
p = (ggplot(df) + aes(x = "cut", y = 'price') + geom_boxplot())
p.draw()
plt.show()
```

```python
p = (ggplot(df) + aes(x = "price") + geom_histogram(bins = 50))
p.draw()
plt.show()
```

```python
df = pd.DataFrame({"x": [1,2,3,4,5], "y": [1,2,1.5, 3, 2.5]})
df
```

```python
p = (ggplot(df) + aes(x = "x", y = "y") + geom_point() + geom_smooth(method = "lm"))
p.draw()
plt.show()
```

```python
df = pd.DataFrame({"x": [1,2,3,4,5,6,7,8,9], "y": [1,2,1.5, 3, 2.5, 3.5, 2.5, 3.5, 4], "group": ["A", "A", "A", "B", "B", "B", "C", "C", "C"]})
df
```

```python
p = (ggplot(df) + aes(x = "x", y = "y") + geom_point() + facet_wrap("group"))
p.draw()
plt.show()
```
